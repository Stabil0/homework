<h2>Задание 1.1</h2>

На основании этих данных необходимо составить максимальное(но не менее пяти) количество гипотез о корреляциях между отдельными столбцами и проверить их. 

При этом при составлении гипотез можно рассматривать не полный df, а именно выборки (например: "женщины старше 40 лет покупают в основном осенью")

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

file_path = 'shopping_trends.csv'
data = pd.read_csv(file_path)
data.head()
```

```python
'''
Гиптоеза 1) Клиенты старшего возраста (старше 40 лет) чаще совершают покупки зимой, в то время как молодёжь 
(до 30 лет) предпочитает покупки весной.
Гипотеза 2) Женщины чаще покупают одежду, а мужчины чаще приобретают обувь.
Гипотеза 3) Товары из категории "Clothing" чаще отправляются с бесплатной доставкой, в то время как 
"Footwear" доставляется через платные сервисы, такие как "Next Day Air".
Гипотеза 4) Покупатели, использовавшие промокоды, оставляют более высокие оценки, чем те, кто их не использовал.
Гипотеза 5) Клиенты с активной подпиской совершают больше покупок в прошлом, чем те, у кого подписка отсутствует.
Гипотеза 6) Осенью и зимой клиенты чаще используют скидки, чем весной и летом.
'''
```

```python

def check_correlation(data, column1, column2, method="chi2"):
    """
    Check the correlation between two categorical variables using Chi-Square or other methods.
    """
    contingency_table = pd.crosstab(data[column1], data[column2])
    if method == "chi2":
        chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
        return {"chi2_stat": chi2, "p_value": p, "dof": dof, "expected": expected}
    return None


# Results storage
results = {}

# 1. Возраст и сезон покупок  
age_bins = pd.cut(data["Age"], bins=[0, 30, 40, 100], labels=["<30", "30-40", ">40"])
season_correlation = check_correlation(data.assign(Age_Group=age_bins), "Age_Group", "Season")
results["Age and Season"] = season_correlation

# 2. Пол и категория товаров
gender_category_correlation = check_correlation(data, "Gender", "Category")
results["Gender and Category"] = gender_category_correlation

# 3. Категория товара и метод доставки
category_shipping_correlation = check_correlation(data, "Category", "Shipping Type")
results["Category and Shipping Type"] = category_shipping_correlation

# 4. Оценка отзыва и использование промокодов)
rating_bins = pd.cut(data["Review Rating"], bins=[0, 3, 4, 5], labels=["Low", "Medium", "High"])
promo_correlation = check_correlation(data.assign(Rating_Group=rating_bins), "Rating_Group", "Promo Code Used")
results["Review Rating and Promo Code Used"] = promo_correlation

# 5. Количество предыдущих покупок и статус подписки
subscription_correlation = check_correlation(data, "Subscription Status", "Previous Purchases")
results["Previous Purchases and Subscription Status"] = subscription_correlation

# 6. Сезон и использование скидок
season_discount_correlation = check_correlation(data, "Season", "Discount Applied")
results["Season and Discount Applied"] = season_discount_correlation

results
```

<h1>Задание 3</h1> 

У нас есть полный каталог всех случаев, когда кто-то проклинал или истекал кровью в фильме Квентина Тарантино. https://github.com/fivethirtyeight/data/tree/master/tarantino

Определите:

* число проклятий, летальных исходов и их соотношение по фильмам 

* частоту употребления конкретных проклятий 

* распределение времени между проклятиями 

```python
import matplotlib.pyplot as plt

data = pd.read_csv("tarantino.csv")

# Задача 1: Число проклятий, летальных исходов и их соотношение
curse_count = data[data["type"] == "word"].groupby("movie").size()
# Если есть данные о смертях, можно добавить:
death_count = data[data["type"] == "death"].groupby("movie").size() if "death" in data["type"].values else pd.Series(0, index=curse_count.index)
curse_death_ratio = curse_count / death_count.replace(0, pd.NA)  # Предотвращаем деление на 0

print("Число проклятий по фильмам:")
print(curse_count)
print("\nЧисло летальных исходов по фильмам:")
print(death_count)
print("\nСоотношение проклятий к летальным исходам:")
print(curse_death_ratio)

# Задача 2: Частота употребления конкретных проклятий
word_frequency = data[data["type"] == "word"].groupby("word").size()
print("\nЧастота употребления проклятий:")
print(word_frequency)

# Задача 3: Распределение времени между проклятиями
curse_times = data[data["type"] == "word"].sort_values(by=["movie", "minutes_in"])
curse_times["time_diff"] = curse_times.groupby("movie")["minutes_in"].diff()
time_differences = curse_times["time_diff"].dropna()

# Построим гистограмму
plt.figure(figsize=(8, 6))
plt.hist(time_differences, bins=10, color='blue', alpha=0.7, edgecolor='black')
plt.title("Распределение времени между проклятиями")
plt.xlabel("Время между проклятиями (минуты)")
plt.ylabel("Частота")
plt.grid(True)
plt.show()
```
